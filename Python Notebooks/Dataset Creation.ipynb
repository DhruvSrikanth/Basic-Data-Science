```python
import pandas as pd
import numpy as np
import random

Start = 0
Size = 50000

coin_toss = ['Heads','Tails']
```

```python
def Rand_Integer (start,end,n):
    """Return a list of random integers"""
    res = list()
    
    for j in range (n):
        res.append(random.randint(start,end))
        
    return res
```

```python
def Rand_Unique_Integer (start,end,n):
    """Return a list of UNIQUE random integers"""
    set_res = set()
    while len(set_res)<n:
        set_res.add(random.randint(start,end))
        
    return set_res
```

```python
def random_decision(feature,options,size):
    """Return a list containing the results drawn through the random algorithm i.e 
    To choose an option to enter data for a feature is  by tossing the coin.If the Coin turns out to be head, then roll a dice with number of faces equal to the number of options a feature can take as given above; In case of a tail enter a predefined option
    """
    Gender_Vector = list()
    Marital_Status_Vector = list()
    Education_Vector = list()
    City_Type_Vector = list()
    for i in range(size):
        coin_toss_res = random.sample(coin_toss,1)
        if (feature == 'Gender'):    
            if (coin_toss_res[0] == 'Heads'):
                faces = options
                faces_res = (random.sample(options,1))[0]
            else:
                faces_res = 'Male'
            Gender_Vector.append(faces_res)
        elif (feature == 'Marital Status'):    
            if (coin_toss_res[0] == 'Heads'):
                faces = options
                faces_res = (random.sample(options,1))[0]
            else:
                faces_res = 'Yes'
            Marital_Status_Vector.append(faces_res)
        elif (feature == 'Education'):    
            if (coin_toss_res[0] == 'Heads'):
                faces = options
                faces_res = (random.sample(options,1))[0]
            else:
                faces_res = 'Ph.d'
            Education_Vector.append(faces_res)
        elif (feature == 'City Type'):    
            if (coin_toss_res[0] == 'Heads'):
                faces = options
                faces_res = (random.sample(options,1))[0]
            else:
                faces_res = 'B'
            City_Type_Vector.append(faces_res)
        else:
            pass
    if (feature == 'Gender'):
        return Gender_Vector
    elif (feature == 'Marital Status'):
        return Marital_Status_Vector
    elif (feature == 'Education'):
        return Education_Vector
    elif (feature == 'City Type'):
        return City_Type_Vector
    else:
        pass
```

```python
def Entity_ID_Generator(Size):
    """Entity IDs are generated and a list containing them is returned"""
    Entity_ID = list()
    
    i=0
    for i in range(Size):
        id = i
        Entity_ID.append(id)
        
    return Entity_ID
```

```python
Entity_ID = Entity_ID_Generator(Size) #Create Entity ID list
```

```python
City_Type_Expenses = {"A":15000,"B":10000,"C":5000,"D":2500} #Options for several features
Gender = ["Female","Male"]
Marital_Status = ["Yes","No"]
Education = ["Not Educated","UG","PG","Ph.d"]
#print(City_Type_Expenses.keys())
#print(City_Type_Expenses.values())
```

```python
Gender_Vector = random_decision('Gender',Gender,Size) #Creating columns for the Dataset
City_Type_Vector = random_decision('City Type',City_Type_Expenses.keys(),Size)
Marital_Status_Vector = random_decision('Marital Status',Marital_Status,Size)
Entity_ID_Vector = Entity_ID
Dependents_Vector = Rand_Integer(Start,5,Size)
Education_Vector = random_decision('Education',Education,Size)
Entity_Income_Vector = Rand_Integer(5000,50000,Size)

Entity_Family_Income_Vector = list()
for i in range(Size):
    income = Entity_Income_Vector[i]
    Total_Income = random.randint(income,500000)
    
    Entity_Family_Income_Vector.append(Total_Income)

Average_Expense_perEntity_PerMonth_Vector = list()
for city in City_Type_Vector:
    if city == "A":
        average_expense = City_Type_Expenses[city]
    elif city == "B":
        average_expense = City_Type_Expenses[city]
    elif city == "C":
        average_expense = City_Type_Expenses[city]
    elif city == "D":
        average_expense = City_Type_Expenses[city]
    else:
        pass
    Average_Expense_perEntity_PerMonth_Vector.append(average_expense)

Rounded_Entity_Income_Vector = list()
for num in Entity_Income_Vector:
    element = round(num,-3)
    Rounded_Entity_Income_Vector.append(element)

Rounded_Entity_Family_Income_Vector = list()
for num in Entity_Family_Income_Vector:
    element = round(num,-3)
    Rounded_Entity_Family_Income_Vector.append(element)
```

```python
DataSet = pd.DataFrame() #Adding columns to the Dataset
DataSet["Entity ID"] = Entity_ID_Vector
DataSet["Gender"] = Gender_Vector
DataSet["Marital Status"] = Marital_Status_Vector
DataSet["Dependents"] = Dependents_Vector
DataSet["Education"] = Education_Vector
DataSet["Entity's Income in INR"] = Rounded_Entity_Income_Vector
DataSet["Entity's Family Income in INR"] = Rounded_Entity_Family_Income_Vector
DataSet["City Type"] = City_Type_Vector
DataSet["Average Expenditure per Entity per Month in INR"] = Average_Expense_perEntity_PerMonth_Vector
```
