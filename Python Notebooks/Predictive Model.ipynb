```python
import pandas as pd
```

```python
Economic_Status_DataSet = pd.read_csv('Economic Status DataSet.csv')
```

```python
Economic_Status_DataSet.drop(columns=["Unnamed: 0"])
```

```python
def scale_down(vector,scaling_factor):
    """Scales the values of a list down by the scaling factor and returns a list with these values"""
    size = len(vector)
    scaled_vector = list()
    for i in range(size):
        scaled_vector.append(vector[i]/scaling_factor)
    
    return scaled_vector
```

```python
Subset = pd.DataFrame()
Dependents_scaled_vector = list()
Family_Income_scaled_vector = list()
Average_Expenditure_scaled_vector = list()
Dependents_scaling_factor = 10
Family_Income_scaling_factor = 1000000
Average_Expenditure_scaling_factor = 100000
Dependents_vector = Economic_Status_DataSet["Dependents"]
Family_Income_vector = Economic_Status_DataSet["Entity's Family Income in INR"]
Average_Expenditure_vector = Economic_Status_DataSet["Average Expenditure per Entity per Month in INR"]

Dependents_scaled_vector = scale_down(Dependents_vector,Dependents_scaling_factor)
Family_Income_scaled_vector = scale_down(Family_Income_vector,Family_Income_scaling_factor)
Average_Expenditure_scaled_vector = scale_down(Average_Expenditure_vector,Average_Expenditure_scaling_factor)
```

```python
Subset["Dependents"] = Dependents_scaled_vector
Subset["Entity's Family Income in INR"] = Family_Income_scaled_vector
Subset["City Type"] = Economic_Status_DataSet["City Type"]
Subset["Average Expenditure per Entity per Month in INR"] = Average_Expenditure_scaled_vector
Subset["Economic Status"] = Economic_Status_DataSet["Economic Status"]
```

```python
Subset
```

```python
#KNN Classifier
```

```python
from sklearn import model_selection
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics

#o. Build a prediction model to find the new entity’s Economic status 

features = ["Dependents","Entity's Family Income in INR","Average Expenditure per Entity per Month in INR"]
X = Subset[features]
Y = Subset["Economic Status"]

# 80% training and 20% test
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, Y, test_size=0.2,random_state=0)

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

training_accuracy = metrics.accuracy_score(y_train,knn.predict(X_train))
testing_accuracy = metrics.accuracy_score(y_test, y_pred)
print("model training accuracy = ",training_accuracy)
print("model testing accuracy = ",testing_accuracy)
report = classification_report(y_test, y_pred)
print(report) #p. Perform metric level analysis of the model
```

```python
#a. The typical shape of the training and testing data
print("shape of training data = ",len(X_train))
print("shape of testing data = ",len(X_test))
```

```python
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```
